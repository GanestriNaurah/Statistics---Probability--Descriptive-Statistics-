```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(data_clean['length'], bins=20, kde=True)
plt.title("Distribusi Panjang Password")
plt.show()

cols = ['num_digits','num_upper','num_lower','num_special']
data_clean[cols].describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(data_clean.drop('password', axis=1).corr(), annot=True, cmap="Blues")
plt.title("Korelasi antar Fitur")
plt.show()
```

###Insight & Kesimpulan Awal
* Pola password umumnya di antara 6 dan 8 karakter.
* Menunjukkan pola data yang mengkhawatirkan
* Penggunaan karakter yang tidak bervariasi

### Potensi Masalah
* Multikolinearitas : Beberapa fitur, seperti **num_lower** dan **num_chars**, sangat berkorelasi yang menyebabkan hasil yang tidak stabil jika menggunakan keduanya dalam model regresi linier.
* Korelasi target yang rendah: Dalam cybersecurity, jika tujuannya adalah memprediksi serangan, beberapa fitur (selain **domain_length**) memiliki korelasi yang rendah. Hal ini berarti fitur-fitur tersebut mungkin tidak cukup kuat untuk membangun model prediksi yang akurat.

### Kesimpulan
* Pengguna cenderung hanya menggunakan huruf kecil (**num_lower**) dibandingkan penggunaan fitur lain seperti huruf besar, angka dan karakter spesial.
* Panjang kata sandi yang cenderung pendek.
*Perlu eksplorasi fitur untuk analisis yang lebih mendalam(misalnya, pola sequensial karakter) agar bisa membangin model yang lebih akurat dalam mengidentifikasi kata sandi yang rentan.
* num_upper & num_special memiliki pengaruh yang sangat kecil dan sering tidak digunakan sehingga kerumitan password menjadi rendah.
